```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(20, 20))
    ax = plt.subplot(111)
    ax.imshow(img, cmap="gray")
```
