```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread("../Computer-Vision-with-Python/DATA/sammy.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

Now, I'll try to get a ROI, that is, the region of interested. In this case, is the sammy face.

```python
roi = img[350:720, 230:710]
plt.imshow(roi)
```

I got it!!

This stage decides which are all edges are really edges and which are not. For this, we need two threshold values, minVal and maxVal. Any edges with intensity gradient more than maxVal are sure to be edges and those below minVal are sure to be non-edges, so discarded. Those who lie between these two thresholds are classified edges or non-edges based on their connectivity. If they are connected to “sure-edge” pixels, they are considered to be part of edges. Otherwise, they are also discarded.

```python
edges = cv2.Canny(roi, 30, 60)
plt.imshow(edges)
```

```python
blurred = cv2.medianBlur(roi, 5)
blurred_edges = cv2.Canny(blurred, 30, 60)
plt.imshow(blurred_edges)
```

```python
blurred_again = cv2.bilateralFilter(blurred, 9, 100, 100)
blurred_edges = cv2.Canny(blurred_again, 30, 60)
plt.imshow(blurred_edges)
```

```python
final_img = cv2.cvtColor(blurred_edges, cv2.COLOR_BGR2RGB)
plt.imshow(final_img)
```
