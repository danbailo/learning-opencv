For default, OpenCV read imgs using BGR channels and Matplotlib plot images using RGB channels

Because of that, we can see the puppy img bluish

```python
# MATPLOTLIB - RGB
# OPENCV - BGR

plt.imshow(img)
```

For that, we can convert the BGR to RGB using the opencv

```python
fix_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```
